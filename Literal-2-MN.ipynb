a. Construya el polinomio por mínimos cuadrados de grado 1 y calcule el error. 


```python
x_i = (0.2,0.3,0.6,0.9,1.1,1.3,1.4,1.6)
y_i = (0.050446,0.098426,0.33277,0.72660,1.0972,1.5697,1.8487,2.5015)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Convertir los datos en arreglos de numpy
x = np.array(x_i)
y = np.array(y_i)

# Realizar el ajuste por mínimos cuadrados 
coeficientes = np.polyfit(x, y, 1)  # Coeficientes del polinomio
polinomio = np.poly1d(coeficientes)  # Crear el polinomio a partir de los coeficientes

# Calcular los valores predichos de y
y_predicción = polinomio(x)

# Calcular el error relativo promedio
error_relativo_promedio = np.mean(np.abs((y - y_predicción) / y))

# Imprimir los resultados
print("Coeficientes del polinomio (a_1, a_0):", coeficientes)
print(f"El polinomio ajustado es: y = {coeficientes[0]:.4f}x  {coeficientes[1]:.4f}")
print(f"Error relativo promedio (ERP): {error_relativo_promedio:.4f}")

# Graficar los puntos y la línea ajustada
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='violet', label='Datos observados')  # Puntos originales
plt.plot(x, y_predicción, color='purple', label=f'Polinomio ajustado\ny = {coeficientes[0]:.4f}x + {coeficientes[1]:.4f}')  # Línea ajustada
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ajuste por mínimos cuadrados con un polinomio de grado 1')
plt.legend()
plt.grid()
plt.show()
```

d. Construya el polinomio por mínimos cuadrados de la forma  y calcule el error. 

```python
x_i = (0.2,0.3,0.6,0.9,1.1,1.3,1.4,1.6)
y_i = (0.050446,0.098426,0.33277,0.72660,1.0972,1.5697,1.8487,2.5015)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Convertir los datos en arreglos de numpy
x = np.array(x_i)
y = np.array(y_i)

# Transformar y tomando el logaritmo natural
log_y = np.log(y)

# Realizar el ajuste por mínimos cuadrados para un polinomio de grado 1 (recta) sobre ln(y)
coeficientes = np.polyfit(x, log_y, 1)  # Ajuste lineal en ln(y)
polinomio = np.poly1d(coeficientes)  # Crear el polinomio a partir de los coeficientes

# Calcular los valores predichos de ln(y)
log_y_predicción = polinomio(x)

# Convertir los valores predichos de ln(y) de vuelta a y usando la exponencial
y_predicción = np.exp(log_y_predicción)

# Calcular el error relativo promedio
error_relativo_promedio = np.mean(np.abs((y - y_predicción) / y))

# Extraer los coeficientes a y b
a = coeficientes[0]  # Coeficiente de x en ln(y) = A + Bx
b = np.exp(coeficientes[1])  # Intercepto, ln(b) = coeficiente[1], por lo que b = exp(coeficiente[1])

# Imprimir los resultados
print(f"Coeficientes del ajuste (a, b): a = {a:.4f}, b = {b:.4f}")
print(f"El polinomio ajustado es: y = {b:.4f}e^{a:.4f}x")
print(f"Error relativo promedio (ERP): {error_relativo_promedio:.4f}")

# Graficar los puntos y la curva ajustada
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Datos observados')  # Puntos originales
x_linea = np.linspace(min(x), max(x), 500)  # Más puntos para suavizar la línea ajustada
y_linea = b * np.exp(a * x_linea)  # Calcular los valores de y ajustados
plt.plot(x_linea, y_linea, color='turquoise', label=f'Curva ajustada\ny = {b:.4f}e^{a:.4f}x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ajuste por mínimos cuadrados (Forma: y = be^ax)')
plt.legend()
plt.grid()
plt.show()
```

```python
x_i = (0.2,0.3,0.6,0.9,1.1,1.3,1.4,1.6)
y_i = (0.050446,0.098426,0.33277,0.72660,1.0972,1.5697,1.8487,2.5015)
```
