```python
peso_promedio = (4800,3700,3400,2800,1900)
porcentaje_de_presentación = (3.1,4.0,5.2,6.4,9.6)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos proporcionados: Peso promedio (en lb) y porcentaje de presentación
peso_promedio = np.array([4800, 3700, 3400, 2800, 1900])  # Pesos en libras
porcentaje_de_presentación = np.array([3.1, 4.0, 5.2, 6.4, 9.6])  # Porcentajes

# Realizar el ajuste por mínimos cuadrados para un polinomio de grado 1 (recta)
coeficientes = np.polyfit(peso_promedio, porcentaje_de_presentación, 1)  # Coeficientes del polinomio
polinomio = np.poly1d(coeficientes)  # Crear el polinomio a partir de los coeficientes

# Calcular los valores predichos de y (porcentaje de presentación)
y_predicción = polinomio(peso_promedio)

# Calcular el error relativo promedio
error_relativo_promedio = np.mean(np.abs((porcentaje_de_presentación - y_predicción) / porcentaje_de_presentación))

# Imprimir los resultados
print("Coeficientes del polinomio (a_1, a_0):", coeficientes)
print(f"El polinomio ajustado es: y = {coeficientes[0]:.4f}x + {coeficientes[1]:.4f}")
print(f"Error relativo promedio (ERP): {error_relativo_promedio:.4f}")

# Graficar los puntos y la línea ajustada
plt.figure(figsize=(8, 6))
plt.scatter(peso_promedio, porcentaje_de_presentación, color='blue', label='Datos observados')  # Puntos originales
plt.plot(peso_promedio, y_predicción, color='orange', label=f'Polinomio ajustado\ny = {coeficientes[0]:.4f}x + {coeficientes[1]:.4f}')  # Línea ajustada
plt.xlabel('Peso Promedio (lb)')
plt.ylabel('Porcentaje de Presentación')
plt.title('Ajuste por mínimos cuadrados: Peso Promedio vs Porcentaje de Presentación')
plt.legend()
plt.grid()
plt.show()
```
