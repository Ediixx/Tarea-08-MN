```python
puntaje_ACT = (28,25,28,27,28,33,28,29,23,27,29,28,27,29,21,28,28,26,30,24)
promedio_puntos = (3.84,3.21,3.23,3.63,3.75,3.20,3.41,3.38,3.53,2.03,3.75,3.65,3.87,3.75,1.66,3.12,2.96,2.92,3.10,2.81)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Convertir los datos en arreglos de numpy
x = np.array(puntaje_ACT)
y = np.array(promedio_puntos)

# Realizar el ajuste por mínimos cuadrados para un polinomio de grado 1
coeficientes = np.polyfit(x, y, 1)  # Coeficientes del polinomio
polinomio = np.poly1d(coeficientes)  # Crear el polinomio a partir de los coeficientes

# Calcular los valores predichos de y
y_predicción = polinomio(x)

# Calcular el error relativo promedio
error_relativo_promedio = np.mean(np.abs((y - y_predicción) / y))

# Imprimir los resultados
print("Coeficientes del polinomio (a_1, a_0):", coeficientes)
print(f"El polinomio ajustado es: y = {coeficientes[0]:.4f}x + {coeficientes[1]:.4f}")
print(f"Error relativo promedio (ERP): {error_relativo_promedio:.4f}")

# Graficar los puntos y la línea ajustada
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='brown', label='Datos observados')  # Puntos originales
plt.plot(x, y_predicción, color='black', label=f'Polinomio ajustado\ny = {coeficientes[0]:.4f}x + {coeficientes[1]:.4f}')  # Línea ajustada
plt.xlabel('Puntaje ACT')
plt.ylabel('Promedio de puntos')
plt.title('Ajuste por mínimos cuadrados Puntaje ACT vs Promedio de puntos')
plt.legend()
plt.grid()
plt.show()
```
